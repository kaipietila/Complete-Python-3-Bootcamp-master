#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
import math
class Line:
    
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
    
    def distance(self):
        return math.sqrt((self.coor1[0]-self.coor2[0])**2 + (self.coor1[1]-self.coor2[1])**2)
    
    def slope(self):
        return ((self.coor1[1]-self.coor2[1]) / (self.coor1[0]-self.coor2[0]))
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 2

Fill in the class 

```python
class Cylinder:
    
    pi = 3.14
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return self.height * self.pi * (self.radius**2)
    
    def surface_area(self):
        return 2*self.pi*(self.radius**2) + 2*self.pi*self.radius*self.height
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
